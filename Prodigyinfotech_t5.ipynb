```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("traffic_accidents.csv")
```

```python
print(df.head())
print(df.info())
```

```python
df.dropna(inplace=True)
```

```python
print(df.columns)
```

```python
df['Road_surface_type'] = df['Road_surface_type'].astype('category')
df['Weather_conditions'] = df['Weather_conditions'].astype('category')
df['Light_conditions'] = df['Light_conditions'].astype('category')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Road_surface_type')
plt.title('Accident Counts by Road Surface Type')
plt.xlabel('Road Surface Type')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Weather_conditions')
plt.title('Accident Counts by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Light_conditions')
plt.title('Accident Counts by Light Conditions (Time of Day)')
plt.xlabel('Light Conditions (Time of Day)')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```
